## AD Anderson - Darling Test

```python
adresults = st.anderson(df['milk'], 'norm')
adresults
```

```python
adresults = st.anderson(df['milk'], 'norm')

statistic = adresults.statistic

H0 = 'H0: Milk Weight of 5 Top Cow Type are Normally Distributed'

HA = 'HA: Milk Weight of 5 Top Cow Type are not Normally Distributed'

if statistic < adresults.critical_values[2]:
    print('The statistical value is {}, and the critical value is {}'.format(statistic, round(adresults.critical_values[2],4))) 
    print()
    print()
    print('H0 fails to reject. {}'.format(H0)) 
else:    
    print('The statistical value is {}, and the critical value is {}'.format(statistic, round(adresults.critical_values[2],4))) 
    print()
    print('H0 rejected. {}'.format(HA)) 
```

## Hypothesis Testing for Milk Weight

```python
#I am confident that 2 best choice for Cargill is Hereford and Swiss
```

```python
Hereford = 44.70
Swiss = 43.10
alpha = 0.05
milky = df['milk'].mean()
sample_size=1000
milk_samples=np.random.choice(df['milk'],sample_size)
```

```python
milk_samples.shape
```

Bileteral (Two-sided) Hypothesis Testing

```python

results1 = ttest_1samp(milk_samples, Hereford)

pvalue1 = results1.pvalue

H0 = 'H0: Mean milk weights of are equal to Hereford mean'

HA = 'HA: Mean milk weights of are not equal to to Hereford mean'

if pvalue1 < alpha:
    print('The Mean milk weights of Hereford is {} kg and mean weight of Mean milk weights is {} kg'.format(Hereford, round(milky,2))) 
    print()
    print('the p-value is {}'.format(pvalue1)) 
    print()
    print('H0 rejected. {}'.format(HA)) 
else: 
    print('The Mean milk weights of Hereford is {} kg and mean weight of Mean milk weights is {} kg'.format(Hereford, round(milky,2))) 
    print()
    print('the p-value is {}'.format(pvalue1)) 
    print()
    print('H0 fails to reject. {}'.format(H0)) 
```

```python
results2 = ttest_1samp(milk_samples, Swiss)

pvalue2 = results2.pvalue

H0 = 'H0: Mean weight of Swiss Milk are equal to mean weight of Milk weight'

HA = 'H0: Mean weight of Swiss Milk are not equal to mean weight of Milk weight'

if pvalue2 < alpha:
    print('The Mean weight of Swiss Milk is {} kg and mean weight of Milk weight is {} kg'.format(Swiss, round(milky,2))) 
    print()
    print('the p-value is {}'.format(pvalue2)) 
    print()
    print('H0 rejected. {}'.format(HA)) 
else: 
    print('The Mean weight of Swiss Milk is {} kg and mean weight of Milk weight is {} kg'.format(Swiss, round(milky,2))) 
    print()
    print('the p-value is {}'.format(pvalue2)) 
    print()
    print('H0 fails to reject. {}'.format(H0)) 
```
